Search Question

Which webpages are most likely to lose search ranking and should be prioritized for optimization?

Decision

Decide whether a webpage should be prioritized for SEO improvements.

Unit of Analysis

One webpage.

Output

A prediction indicating whether the webpage is likely to lose search performance or require optimization.

Action

If the model predicts a page is at risk, the SEO or content team updates the content, improves keywords, fixes technical issues, or refreshes the page.

False Negative

An important page loses ranking and traffic because no action was taken.
